```python
import matplotlib.pyplot as plt
%matplotlib inline
```

<h1>normal plot</h1>

```python
x=[1,2,3,4,5,6]
y=[4,7,8,9,0,8]
```

```python
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(x,y,color='green',linewidth=5,linestyle='solid')
# dashed,dotted,dashdot,solid
```

```python
plt.plot(x,y,'g+')
# g means green and + means + sign
```

```python
plt.plot(x,y,color='green',linestyle='solid',alpha=0.5)
# alpha using for transpercy
```

<h1>uses of legend</h1>

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,53,54,55,56]
min_t=[41,42,43,44,45,46,47]
avg_t=[45,46,47,48,49,50,51]
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='upper right',shadow=True,fontsize='small')
plt.grid()
# grid use for grid line
# loc= upper right or best or any direction can be ,best use for space plce
```

<h1>bar chart<h1>

```python
import numpy as np
```

```python
company=['google','amzn','msft','fb']
revenue=[90,136,89,60]
profit=[41,42,40,46]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.ylabel('revenue')
plt.xticks(xpos,company)
plt.bar(xpos,revenue,label='revenue')
plt.bar(xpos,profit,label='profit')
plt.legend()
```

```python
plt.ylabel('revenue')
plt.xticks(xpos,company)
plt.bar(xpos+0.2,revenue,label='revenue',width=0.4)
plt.bar(xpos-0.2,profit,label='profit',width=0.4)
plt.legend()
```

```python
# horizontal bar chart
plt.ylabel('revenue')
plt.xticks(xpos,company)
plt.barh(xpos+0.2,revenue,label='revenue')
plt.barh(xpos-0.2,profit,label='profit')
plt.legend()
```

<h1>histogram</h1>

```python
blood_sugar=[113,85,89,90,45,85,110,135,77,82]
plt.hist(blood_sugar,bins=3,rwidth=0.95)
```

```python
blood_sugar=[113,85,89,90,45,85,110,135,77,82]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95)
# 80 to 100 & 100 to 125 &125 to 150
```

<h1>piechart</h1>

```python
exp_value=[1400,600,300,410,259]
exp_label=['homerent','food','phonebill','car','others']
plt.axis('equal')
plt.pie(exp_value,labels=exp_label)
plt.show()

```

```python
plt.axis('equal')
plt.pie(exp_value,labels=exp_label,autopct='%0.1f%%')
plt.show()
```

```python
# explode
plt.axis('equal')
plt.pie(exp_value,labels=exp_label,autopct='%0.1f%%',explode=[0,0.1,0,0,0])
plt.show()
```

```python
# rotating position
plt.axis('equal')

plt.pie(exp_value,labels=exp_label,autopct='%0.1f%%',startangle=180)
plt.show()
```

```python
# save chart
plt.axis('equal')

plt.pie(exp_value,labels=exp_label,autopct='%0.1f%%',startangle=180)
plt.show()
plt.savefig('piechart.pdf',bbox_inches='tight',transparent=True)
```

<h1>subplot</h1>


```python
import numpy as np
x=np.array([0,1,2,3])
y=np.array([3,4,5,7])
plt.subplot(1,2,1)
# 1 is row 2 is coloumn & last is index
plt.plot(x,y)
# plot2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
# 1is row 2 is coloumn and last 2 is index
plt.plot(x,y)
plt.show()
```

```python
# title & super title
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```
